# ***1-	Introducción***

El análisis de datos en el sector automotriz es fundamental para comprender las tendencias del mercado y los factores que influyen en los precios de los vehículos usados. En este estudio, se utiliza un dataset que proporciona información sobre automóviles usados, el mismo fue obtenido del sitio Kaggle: https://www.kaggle.com/datasets/thedevastator/uncovering-factors-that-affect-used-car-prices/data .
Este conjunto de datos contiene variables que ofrecen una visión detallada de cada vehículo y su contexto en el mercado de segunda mano.

Entre las variables incluidas en el dataset se encuentran:

- Información de búsqueda (dateCrawled): Esta variable indica la fecha y hora en que se buscó cada automóvil, lo que puede proporcionar información sobre la temporalidad de los datos.
-	Nombre del vehículo (name): Describe el nombre del automóvil.
-	Tipo de vendedor (seller): Clasifica al vendedor como particular o concesionario, lo que puede influir en la percepción del comprador sobre la confiabilidad y la calidad del servicio postventa.
-	Tipo de oferta (offerType): Indica el tipo de oferta, que puede ser de venta, intercambio, alquiler, entre otros, lo que proporciona información adicional sobre las condiciones de la transacción.
-	Precio (price): Precio del automóvil, una variable crucial que influye en la decisión de compra.
-	Prueba AB (abtest): Esta variable indica si se realizó una prueba al vehículo.
-	Tipo de vehículo (vehicleType): Describe el tipo de vehículo, como SUV, sedán, coupé, etc., lo que ayuda a segmentar el mercado según las preferencias de los compradores.
-	Año de registro (yearOfRegistration): Año en que se fabricó el automóvil, lo que puede afectar su valor de mercado debido a la antigüedad y la disponibilidad de tecnologías más recientes.
-	Tipo de caja de cambios (gearbox): Transmisión manual o automática, una característica importante que puede influir en las preferencias de compra de los consumidores.
-	Potencia (powerPS): Representa la potencia del motor en caballos de fuerza (PS), un factor determinante en el rendimiento y la velocidad del automóvil.
-	Modelo (model): Indica el modelo específico del automóvil, lo que permite realizar comparaciones y análisis detallados entre diferentes modelos y marcas.
-	Kilometraje (kilometer): Muestra la distancia recorrida por el automóvil en kilómetros, una variable importante que afecta su estado y valor de mercado.
-	Mes de registro (monthOfRegistration): Indica el mes en que se fabrico el automóvil por primera vez, lo que puede proporcionar información sobre la temporada de compra y venta de vehículos.
-	Tipo de combustible (fuelType): Describe el tipo de combustible utilizado por el automóvil, como Nafta, diésel, eléctrico, etc, lo que puede influir en los costos operativos y la demanda del mercado.
-	Marca (brand): Representa la marca del automóvil, un factor importante que puede influir en la percepción de calidad, confiabilidad y prestigio del vehículo.
-	Daños no reparados (notRepairedDamage): Indica si el automóvil tiene algún daño que no ha sido reparado, lo que puede afectar significativamente su valor y la disposición de los compradores a pagar un precio determinado.
-	Fecha de creación (dateCreated): Día / mes / año en que se publico el automóvil.
-	Número de fotografías (nrOfPictures): Número de fotografías disponibles.
-	Código postal (postalCode): Código postal de la ubicación del vendedor, lo que puede proporcionar información geográfica relevante para análisis regionales y tendencias de mercado.
-	Última visualización (lastSeen): Indica la fecha en que el automóvil fue visto por última vez en la web.

El análisis de este conjunto de datos permite explorar las relaciones entre estas variables y comprender cómo influyen en los precios de los automóviles usados. Al investigar estas relaciones, es posible identificar patrones, tendencias y factores clave que afectan la valoración y la demanda en el mercado de segunda mano. Esto puede proporcionar información valiosa tanto para los compradores como para los vendedores, permitiéndoles tomar decisiones más informadas y estratégicas en sus transacciones comerciales.

Los terminos kleinwagen y limousine, dado la conceptualidad de lo que engloban en la industria automotriz alemana no se traduciran. A saber:

Son vehículos de tamaño pequeño a mediano diseñados principalmente para el transporte urbano y suelen tener capacidades de pasajeros de cuatro a cinco personas. Ejemplos comunes de Kleinwagen son el Volkswagen Polo, el Ford Fiesta y el Toyota Yaris.

En Europa, "Limousine" se utiliza para describir un tipo de carrocería de automóvil que generalmente se refiere a sedanes de tamaño grande o sedanes de lujo. Estos vehículos suelen ser más grandes que los Kleinwagen y están diseñados para ofrecer un mayor espacio interior y comodidad para los pasajeros. Ejemplos comunes de Limousine son el Mercedes-Benz Clase E, el BMW Serie 5 y el Audi A6.

### ***Se procede a eliminar Outliers***

Como se observa en el analisis de la columna Price, se pueden ver valores muy atipicos en los datos.
Dicho esto es que se propone hacer el corte por los rangos intercuartilicos.

Variable Price

```python
# 1. Calcular cuartiles Q1 y Q3
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

# 2. Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# 3. Definir límites para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 4. Eliminar outliers
df = df[(df['price'] >= limite_inferior) & (df['price'] <= limite_superior)]
```

Variable yearOfRegistration

```python
Q1 = df['yearOfRegistration'].quantile(0.25)
Q3 = df['yearOfRegistration'].quantile(0.75)

# 2. Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# 3. Definir límites para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 4. Eliminar outliers
df = df[(df['yearOfRegistration'] >= limite_inferior) & (df['yearOfRegistration'] <= limite_superior)]
```

Variable powerPS

```python
Q1 = df['powerPS'].quantile(0.25)
Q3 = df['powerPS'].quantile(0.75)

# 2. Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# 3. Definir límites para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 4. Eliminar outliers
df = df[(df['powerPS'] >= limite_inferior) & (df['powerPS'] <= limite_superior)]
```

```python
import statsmodels.api as sm
def forward_selection(data, target, significance_level=0.01):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

```python
# Seleccionar características utilizando forward selection
selected_features = forward_selection(X, y)

# Mostrar las características seleccionadas
print("****************************************************************************************************")
print("Características seleccionadas:", selected_features)
print("****************************************************************************************************")
```

### MAE

```python
from sklearn.metrics import mean_absolute_error
print("**********************")
print("MAE",mean_absolute_error(y_test,y_pred))
print("**********************")
```

El MAE (Mean Absolute Error) evalua la precisión de un modelo predictivo. Indica el promedio de las diferencias absolutas entre las predicciones del modelo y los valores reales.

Para este caso, un MAE de 1813.94 significa que, en promedio, las predicciones del modelo tienen una diferencia absoluta de aproximadamente 1813.94 unidades (en la misma escala que la variable Price) en comparación con los valores reales. En otras palabras, en promedio, las predicciones están desviadas de los valores reales en alrededor de 1813.94 unidades (+/-).

### MSE

```python
print("**********************")
print("MSE",mean_squared_error(y_test,y_pred))
print("**********************")
```

El MSE (Mean Squared Error) El MSE calcula el promedio de los cuadrados de las diferencias entre las predicciones del modelo y los valores reales.

Para el caso, un MSE de 6182330.30 significa que, en promedio, el cuadrado de las diferencias entre las predicciones del modelo y los valores reales es de aproximadamente 6182330.30 unidades al cuadrado (en la misma escala que tus datos de precios). En otras palabras, las diferencias individuales entre las predicciones y los valores reales se elevan al cuadrado, se promedian y ese valor resultante es el MSE.

Como el MSE se calcula elevando al cuadrado las diferencias, es más sensible a los errores grandes en comparación con el MAE. Un MSE más bajo indica una mejor precisión del modelo, ya que significa que las predicciones del modelo tienen diferencias más pequeñas en promedio con los valores reales. Por otro lado, un MSE más alto indica una menor precisión, ya que las diferencias entre las predicciones del modelo y los valores reales tienden a ser mayores.

### RMSE

```python
print("**********************")
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
print("**********************")
```

El RMSE (Root Mean Squared Error) Es similar al MSE, pero en lugar de tomar el promedio de las diferencias al cuadrado entre las predicciones del modelo y los valores reales, el RMSE toma la raíz cuadrada de este valor promedio. Esto proporciona una métrica de error que está en la misma escala que los datos originales.

Para el caso, un RMSE de 2486.43 indica que, en promedio, las diferencias entre las predicciones del modelo y los valores reales son de aproximadamente 2486.43 unidades en la misma escala que los datos de precios.

El RMSE es una medida útil porque tiene la misma unidad que la variable objetivo (en este caso, el precio), lo que facilita su interpretación. 
Un RMSE más bajo indica una mejor precisión del modelo, ya que significa que las predicciones del modelo tienen diferencias más pequeñas en promedio con los valores reales. 
Por otro lado, un RMSE más alto indica una menor precisión, ya que las diferencias entre las predicciones del modelo y los valores reales tienden a ser mayores.

### RMSLE

```python
print("**********************")
print("RMSE",np.log(np.sqrt(mean_squared_error(y_test,y_pred))))
print("**********************")
```

un RMSLE de 7.8186 indica que las diferencias relativas entre las predicciones del modelo y los valores reales son grandes. Esto sugiere que el modelo puede no estar capturando bien la variabilidad de los datos o que hay características importantes que no se están teniendo en cuenta en el modelo. En general, se busca minimizar el RMSLE, al igual que otras métricas de error, para mejorar la precisión del modelo.

### R2

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print("**********************")
print("R2:", r2)
print("**********************")
```

El coeficiente de determinación, denotado como R2 , es una medida estadística que indica qué tan bien se ajustan los valores observados a los valores predichos por el modelo de regresión. Es un valor que varía entre 0 y 1, donde 1 indica un ajuste perfecto y 0 indica que el modelo no explica nada de la variabilidad de los datos.

Para el caso, un R2 de aproximadamente 0.6044 significa que ***el modelo de regresión explica alrededor del 60.44% de la variabilidad de los datos de la variable objetivo.*** Esto indica que el modelo tiene un ajuste moderado, pero aún hay una cantidad significativa de variabilidad que no se está capturando.
